```python
import glob
import shutil
import os
```

```python
root_dir = '/scratch3/users/liutianyang/katcali_pipeline/'
level = 'level4/'
src = 'py_results/BOX13_20251015_090000/'
name_match = 'F_*_ch_filter_final.png'
dst = 'py_results/BOX13_20251015_090000_ch_filter_final/'
fname_start = 2

src_dir = os.path.join(root_dir, level, src)
dst_root = os.path.join(root_dir, level, dst)
os.makedirs(dst_root, exist_ok=True)
```

```python
pattern = os.path.join(src_dir, name_match)
files = glob.glob(pattern)
print(files[0])
```

```python
for i, f in enumerate(files):
    filename = os.path.basename(f)
    fname = filename[fname_start:fname_start+10]
    if not i % 100:
        print(i, fname)

    dst_dir = os.path.join(dst_root, fname)
    os.makedirs(dst_dir, exist_ok=True)

    shutil.copy2(f, dst_dir)
```
